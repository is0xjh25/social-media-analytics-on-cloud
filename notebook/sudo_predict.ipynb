```python
# population
marital_status = pd.read_csv('../data/raw/sudo/abs_2021census_Social_Marital_Status_by_Age_by_Sex.csv')
marital_status.columns
population = marital_status[['gccsa_code_2021',' p_tot_total']]
population = population.rename(columns={' p_tot_total': 'population'})
```

```python
# marital_status
marital_status = marital_status[['gccsa_code_2021', ' p_tot_not_married']]
marital_status = marital_status.rename(columns={' p_tot_not_married': 'p_tot_not_married'})
```

```python
# higher education
education = pd.read_csv("../data/raw/sudo/abs_2021census_Selected_Labour_Force_Education_and_Migration_Characteristics_by_Sex.csv")
education.columns
education = education.rename(columns=lambda x: x.replace(' ', ''))
education.columns
education['higher_edu'] = education['non_sch_qual_bchelr_degree_p'] + education['non_sch_qual_postgrad_dgre_p'] + education['non_sch_qual_gr_dip_gr_crt_p']
higher_education = education[['gccsa_code_2021', 'higher_edu']]
```

```python
# employment
employment = education [['gccsa_code_2021','percnt_labforc_prticipation_p', 'percent_unem_loyment_p']]
employment
```

```python
# other all in median dataset
median = pd.read_csv('../data/raw/sudo/abs_2021census.csv')
median = median.rename(columns=lambda x: x.replace(' ', ''))
median.columns
```

```python
corelation_df = death_rate.merge(population,how="inner", left_on="gcc", right_on="gccsa_code_2021").drop(columns="gcc").merge(marital_status, how="inner", on= "gccsa_code_2021").merge(higher_education, how="inner", on= "gccsa_code_2021").merge(employment, how="inner", on= "gccsa_code_2021").merge(median, how="inner", on= "gccsa_code_2021")
corelation_df.to_csv("../data/curated/corelation_df.csv")
```
